```python
# Searching for images
# Turn on the "Image Search" for the custom search engine in cse setup
# This is only getting the first 10 results
result_img = resource.list(q='pepsi', cx = '4583f9d92a851dd9c', searchType='image').execute()
for item in result_img['items']:
  print(item['title'], item['link'])
```

```python
# Getting the next results as well
images = []
for i in range(1, 60, 10):
  result_n = resource.list(q='pepsi', cx = '4583f9d92a851dd9c', searchType='image', start=i).execute()
  images += result_n['items']

# Viewing the results  
for item in images:
  print(item['title'], item['link'])
print(len(images))
```

```python
# Getting the next results as well
items = []

for i in range(0, 60, 10):
  results_n = resource.list(q='pepsi', cx = '4583f9d92a851dd9c', start=i).execute()
  items += result_n['items']

for item in items:
  print(item['title'], item['link'])
```

```python
book = xlsxwriter.Workbook("Data.xlsx")
sheet = book.add_worksheet("first_sheet")

for index, entry in enumerate(items):
  sheet.write(index+1, 0, str(index))
  sheet.write(index+1, 1, entry["title"])
  sheet.write(index+1, 2, entry["link"])

book.close()
```
