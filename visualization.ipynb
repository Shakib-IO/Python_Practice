````python
# Data
from typing import List
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('default')


def plot_data_function(model_name: List[str], acc: List[List[float]], attack: List[int], fig_title:str) -> None:

  colors = ['blue', 'green', 'brown', 'purple', 'orange', 'magenta', 'gray', 'lime', 'red']

  # Create figure and axis objects
  fig, ax = plt.subplots(figsize=(10, 8))

  markers = ['*', 's', '^', 'D', 'v', '>', 'p', 'd', 'o']

  # Plot the line graphs for accuracy vs. attack strength for each model
  for i, model in enumerate(model_name):
      acc_values = [row[i] for row in acc]
      ax.plot(attack, acc_values, marker=markers[i], label=model, color=colors[i], fillstyle='full', markersize = 6.0)

  # Set the labels, title, and legend
  ax.set_xlabel('Perturbation Rate (%)', fontsize=20)
  ax.set_ylabel('Test Accuracy (%)', fontsize=20)
  # ax.set_title('Cora', fontsize=20)
  ax.set_xticks(attack)

  ax.tick_params(axis='x', labelsize=20)  # Increase x-axis label fontsize
  ax.tick_params(axis='y', labelsize=20)  # Increase y-axis label fontsize
  ax.legend(fontsize=15, loc='lower left') # Remove legend background 'frameon=False'
  ax.grid(True, linestyle='--', alpha=0.5, linewidth=0.4, color='black')

  fig.suptitle(f'\n{fig_title}', fontsize=20, y=0)
  # Display the graph
  # plt.grid(True)
  plt.tight_layout()
  plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
sns.set(font_scale=1.2)
from matplotlib import rcParams

plt.rcParams['figure.figsize'] = (7, 5)
rcParams.update({'figure.autolayout': True})
rcParams['pdf.fonttype'] = 42

def h(lambd, s):
    return 1 / ((1 + s) * lambd - s * lambd**2)

# Values of s to consider
s_values = [0.1, 0.2, 0.5, 0.7]

# Lambda values
lambd = np.linspace(0.001, 2, 100)
fig, ax = plt.subplots()

# Plot h(lambda) for each value of s, but only for lambd >= 1
for s in s_values:
    h_values = h(lambd, s)
    # Filter values to be within the specified range
    mask = lambd > 1.01
    ax.plot(lambd[mask], h_values[mask], label=f's = {s}')

# Add labels and legend
ax.set_xlabel('$\lambda$')
ax.set_ylabel('$h_s(\lambda)$')
ax.legend(loc='upper left')

# Set x-axis limit
ax.set_xlim(1.0, 2.05)

# Set y-axis ticks
ax.set_yticks([0.0, 0.5, 1.0, 1.5, 2.0])

# Show the plot
ax.grid(True)
plt.show()

fig.savefig('Shakib_Filter.pdf', format='PDF', bbox_inches='tight')
````

### Bar Plot

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
sns.set(font_scale=1.2)
from matplotlib import rcParams
from typing import List, Optional
# plt.style.use('ggplot')

plt.rcParams['figure.figsize'] = (7,5)
rcParams.update({'figure.autolayout': True})
rcParams['pdf.fonttype'] = 42


"""
Template Colors:
colors = ['#FF5733', '#33FF57', '#3391FF', '#FF33B0', '#FFD700', '#4B0082', '#33FFD3', '#808080', '#800000']
"""

def plot_bar2(model_name: List[str], acc: List[List[float]], attack: List[int], fig_name: str, colors: Optional[List[str]] = ['#3391FF', '#9975c0']) -> None:

    # Create figure and axis objects
    fig, ax = plt.subplots(figsize=(7, 5))

    # Set the width of the bars
    bar_width = 0.4

    # Calculate positions for the bars
    positions = np.arange(len(attack))

    # Plot the bar chart for accuracy vs. attack strength for each model
    for i, model in enumerate(model_name):
        acc_values = acc[i]
        bars = ax.bar(positions + i * bar_width, acc_values, bar_width, label=model, color=colors[i])

    # for bar, acc_value in zip(bars, acc_values):
    #         height = bar.get_height()
    #         ax.text(bar.get_x() + bar.get_width() / 2, height, f'{acc_value:.2f}', ha='center', va='bottom', fontsize=10)

    # Set the labels, title, and legend
    ax.set_xlabel('Number of perturbations', fontsize=16)
    ax.set_ylabel('Test Accuracy (%)', fontsize=16)
    ax.set_xticks(positions + 0.2 * bar_width)
    ax.set_xticklabels(attack)

    ax.tick_params(axis='x', labelsize=16)  # Increase x-axis label fontsize
    ax.tick_params(axis='y', labelsize=16)  # Increase y-axis label fontsize

    # Customize the legend position
    ax.legend(fontsize=10, loc='upper right')
    ax.set_ylim(50, 90)
    # ax.yaxis.grid(True, linestyle='--', alpha=0.5, linewidth=0.2, color='white')
    ax.xaxis.grid(False)
    ax.yaxis.grid(True)
    # ax.grid(False, linestyle='--', alpha=0.4, linewidth=0.1, color='black')

    # Save the figure and show
    plt.tight_layout()
    plt.show()
    fig.savefig(f'{fig_name}.pdf', format='PDF', bbox_inches='tight')
````

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
sns.set(font_scale=1.2)
from matplotlib import rcParams
from typing import List, Optional
plt.style.use('default')

plt.rcParams['figure.figsize'] = (7,5)
rcParams.update({'figure.autolayout': True})
rcParams['pdf.fonttype'] = 42

def parameter_sensitivity_analysis01(alpha_values:List[List[float]], acc_values:List[List[float]], attack_values:List[int], fig_name):
    colors = ['#F56038', '#309975','#5f59f7', '#FFD700', '#4B0082', '#33FFD3', '#808080', '#800000']
    markers = ['x', 'D', 'v', '^', 'P', 's', '*', 'o', 'h']

    # Create figure and axis objects
    fig, ax = plt.subplots(figsize=(8, 6))

    # Plot the line graphs for accuracy vs. alpha for each model
    for i, alpha_set in enumerate(alpha_values):
        ax.plot(alpha_set, acc_values[i], marker=markers[i], label=f'Model {i+1}', color=colors[i], fillstyle='full', markersize=6.5)

    # Set the labels, title, and legend
    # ax.set_xlabel('$ \alpha $', fontsize=16)
    # ax.set_xlabel('α', fontsize=16)
    ax.set_xlabel(r'$\alpha$', fontsize=16)
    ax.set_ylabel('Test Accuracy (%)', fontsize=16)
    ax.set_xticks(alpha_values[0])
    ax.tick_params(axis='x', labelsize=14)
    ax.tick_params(axis='y', labelsize=14)
    ax.legend(fontsize=10, loc='upper left', bbox_to_anchor=(0.005, 0.005), ncol=1)
    ax.xaxis.grid(False)
    ax.yaxis.grid(True)

    # Add legend with attack values
    legend_labels = [f'{attack}%' for attack in attack_values]
    ax.legend(legend_labels, fontsize=10, title_fontsize=10, loc='upper right', bbox_to_anchor=(0.005, 0.005))

    # Grid and show plot
    ax.grid(True, linestyle='--', alpha=0.4, linewidth=0.1, color='while')
    plt.tight_layout()
    plt.show()
    # fig.savefig(f'{fig_name}.pdf', format='PDF', bbox_inches='tight')
````
