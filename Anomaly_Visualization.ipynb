```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
sns.set(font_scale=1.2)
from matplotlib import rcParams
from typing import List
plt.style.use('default')

plt.rcParams['figure.figsize'] = (7,5)
rcParams.update({'figure.autolayout': True})
rcParams['pdf.fonttype'] = 42

def parameter_s(datasets: List[str], alpha_values: List[List[float]], auc_values: List[List[float]], fig_name):
    colors = ['#F56038', '#309975', '#5f59f7', '#FFD700', '#4B0082', '#33FFD3', '#808080', '#800000']
    markers = ['x', 'D', 'v', '^', 'P', 's', '*', 'o', 'h']

    # Create figure and axis objects
    fig, ax = plt.subplots(figsize=(7, 5))

    # Plot the line graphs for accuracy vs. alpha for each dataset
    for i, dataset in enumerate(datasets):
        ax.plot(alpha_values[i], auc_values[i], marker=markers[i], label=dataset, color=colors[i], fillstyle='full', markersize=10)

    # Set the labels, title, and legend
    ax.set_xlabel(r'$\alpha$', fontsize=20)
    ax.set_ylabel('AUC', fontsize=20)

    ax.set_xticks(alpha_values[0])
    ax.tick_params(axis='x', labelsize=16)
    ax.tick_params(axis='y', labelsize=16)

    ax.legend(fontsize=12, loc='lower right')



    # Grid and show plot
    ax.grid(False, linestyle='--', alpha=0.3, linewidth=0.1, color = 'black')
    ax.xaxis.grid(True)
    ax.yaxis.grid(True)
    plt.tight_layout()
    plt.show()
    # fig.savefig(f'{fig_name}.pdf', format='PDF', bbox_inches='tight')
```
