```python
uniques = list(set(dfx['sku']))

skus_with_no_comment = []
total_rows_with_this_sku = []
rows_with_no_comment_under_this_sku = []
rows_with_comments_under_this_sku = []
percentage_of_no_comment = []
highest_date_with_this_sku = []
lowest_date_with_this_sku = []
max_date_with_nocom = []
min_date_with_nocom = []


def create_dataframe():

  df1 = pd.DataFrame(list(zip(skus_with_no_comment, total_rows_with_this_sku, rows_with_no_comment_under_this_sku, 
                              rows_with_comments_under_this_sku, percentage_of_no_comment, highest_date_with_this_sku, lowest_date_with_this_sku, 
                              max_date_with_nocom, min_date_with_nocom)), 
                     columns = ["skus_with_no_comment", "total_rows", "rows_with_no_comment", 
                                "rows_with_comments", "no_comment_percentage", "highest_date", "lowest_date", "max_date_with_nocom", "min_date_with_nocom"])

  return df1



def find_data(sku, df):

  df_n = df.loc[df['sku'] == sku].sort_values('date_key')
  
  # if df_n['total_comment_num'].isna().sum() > 0 or df_n['total_comment_num'] == 0.0:
  if ((df_n['total_comment_num'].isna().sum() > 0) | (df_n['total_comment_num'] == 0.0)).any():

    skus_with_no_comment.append(sku)
    total_rows_with_this_sku.append(len(df_n))
    rows_with_no_comment = len(df_n[(df_n['total_comment_num'].isna()) | (df_n['total_comment_num'] == 0.0)])
    rows_with_no_comment_under_this_sku.append(rows_with_no_comment)
    rows_with_comments_under_this_sku.append(len(df_n) - rows_with_no_comment)
    percentage_of_no_comment.append((df_n['total_comment_num'].isna().sum() / len(df_n)) * 100)
    highest_date_with_this_sku.append(df_n.loc[dfx['sku'] == sku, 'date_key'].max().date())
    lowest_date_with_this_sku.append(df_n.loc[dfx['sku'] == sku, 'date_key'].min().date())
    max_date_with_nocom.append(df_n.loc[(dfx['sku'] == sku) & (dfx['total_comment_num'].isna()),  'date_key'].max().date())
    min_date_with_nocom.append(df_n.loc[(dfx['sku'] == sku) & (dfx['total_comment_num'].isna()),  'date_key'].min().date())
  

  return create_dataframe()


def get_data(i, f):

  skus_with_no_comment.clear()
  total_rows_with_this_sku.clear()
  rows_with_no_comment_under_this_sku.clear()
  rows_with_comments_under_this_sku.clear()
  percentage_of_no_comment.clear()
  highest_date_with_this_sku.clear()
  lowest_date_with_this_sku.clear()
  max_date_with_nocom.clear()
  min_date_with_nocom.clear()
  

  for s in uniques[i:f]:
    df = find_data(s, dfx)

  return df
```

# 4. The function to get all the rows for which we can calculate tcd_values.
- As input, the function takes 
  > 1. an sku, 
  > 2. a random date, 
  > 3. the database
  > 4. the value of m.
- The function returns all the rows for which the tcds can be calculated for a given sku and a given date.
- We can easily calculate for how many rows it's possible to calculate the tcd values with this function.

```python
def good_data(sku, date, df, m):
  
  dates, comments, tc_deltas, tcds, day_diffs, indexes, idxs = [], [], [], [], [], [-2, -1], []


  date = datetime.strptime(date, '%Y-%m-%d').date()
  df_n = df.loc[df['sku'] == sku].sort_values('date_key')
  df_n = df_n[df_n['total_comment_num'].notna()]
  df_n = df_n.drop(['product_name', 'monthly_sales_qty', 'total_sales_qty'], axis = 1)
  df_n["day_diff"] = abs(df_n['date_key'].dt.date - date).dt.days
  
  current_day_diff = 0

  df_n = df_n.loc[(abs(current_day_diff - df_n['day_diff']) > m) | (df_n['day_diff'] == 0)].reset_index()



  idx = 0
  comment_on_first_row = df_n['total_comment_num'].iloc[0]



  highest_date = df_n['date_key'].max().date()
  lowest_date = df_n['date_key'].min().date()


# For loop here
  for idx, r in df_n.iterrows():

    if lowest_date <= date <= highest_date:
      try:
        idx = df_n['day_diff'].loc[abs(current_day_diff - df_n['day_diff']) > m].idxmin()

        if idx < indexes[-1]:
          df_n.drop(df_n.index[0:idx+1], inplace=True)

        indexes.append(idx)
        
        
        current_day_diff = df_n['day_diff'][idx]
        closest_date = df_n['date_key'][idx].date()
        comment_on_closest_date = df_n['total_comment_num'][idx]
        df_n = df_n.drop(idx)


        if (df_n.date_key == pd.Timestamp(date)).any():
          comment_on_given_date = df_n.loc[df_n['date_key'] == str(date), 'total_comment_num'].item()
        else:
          comment_on_given_date = comment_on_first_row
        

        tc_delta = abs(abs((date - pd.to_timedelta(30, unit='d')) - closest_date).days)
        tcd = abs(comment_on_closest_date - comment_on_given_date)


        dates.append(closest_date)
        comments.append(comment_on_closest_date)
        tc_deltas.append(tc_delta)
        tcds.append(tcd)
        day_diffs.append(current_day_diff)
        idxs.append(idx)
        
        date = closest_date


      except (ValueError, KeyError) as e:
        pass


      out = {"date": dates,
             "comment": comments,
             "day_diff": day_diffs,
             "tc_delta": tc_deltas,
             "tcd": tcds,
             "sku": sku,
             "idxs": idxs
            }



  return pd.DataFrame(out).set_index(['idxs', 'sku'])


good_data(100005855583, "2020-10-09", dfx, 50)
```
